```python
import numpy as np
import matplotlib.pyplot as plt
n = 1000; u = 200; sigma = 5
np.random.seed(12345679)
X = np.random.normal(u, sigma, n)
hist,bins = np.histogram(X,35) 
mid_bins = [(bins[i-1]+bins[i])/2 for i in range(1,len(bins))]
plt.bar(mid_bins,hist)
plt.title('Histogram of X')
plt.show()
result = []
sum = 0
for i in range(1,len(hist)+1):
    sum += hist[i-1]
    result.append(sum)
CDF = [result[i-1]/n for i in range(1,len(result)+1)]
plt.plot(mid_bins,CDF)
plt.title('CDF')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = 100; u = 200; sigma = 5
np.random.seed(12345679)
X = np.random.normal(u, sigma, n)
hist,bins = np.histogram(X,n)
mid_bins = [(bins[i-1]+bins[i])/2 for i in range(1,len(bins))]
Xi = [1/(sigma * np.sqrt(2 * np.pi))*np.exp( - (x - u)**2 / (2 * (sigma)**2) ) for x in mid_bins]
result = []
sum = 0; k = 1
for i in range(1,len(Xi)+1):
    sum += Xi[i-1]
    result.append(sum)
    if (i == len(Xi)):
        k = 1/result[i-1]
CDF = [result[i-1] * k for i in range(1,len(result)+1)]
plt.bar(mid_bins,np.multiply(Xi,k))
plt.title('Histogram of X')
plt.show()
plt.plot(mid_bins,CDF)
plt.title('CDF')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = 100; u = 200; sigma = 25
np.random.seed(12345679)
X = np.random.normal(u, sigma, n)
hist,bins = np.histogram(X,n) 
mid_bins = [(bins[i-1]+bins[i])/2 for i in range(1,len(bins))]
CDF = []
sum = 0; P = 0; t = 0
for i in range(1,len(hist)+1):
    sum += hist[i-1]/n
    CDF.append(sum)
    #print(CDF)
    if (mid_bins[i-1] >= u - sigma and mid_bins[i-1] <= u + sigma):
        t += 1
        P += CDF[i-1]
P /= t
print (P)
```
