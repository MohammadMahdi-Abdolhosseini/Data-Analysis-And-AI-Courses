```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r"migrants.csv")
dfdrop = df.drop(columns=['percentarea','ccode'])
country =list(dfdrop['country'])
number1000 = list(dfdrop['number1000'])
number1000_sorted = np.argsort(number1000)
lenght = len(number1000_sorted)
L = number1000_sorted[lenght - 10 : lenght]
max_country=[]
max_number1000_sorted=[]
for i in L:
    max_country.append(country[i])
    max_number1000_sorted.append(number1000[i])
plt.figure(num=1,figsize=(15,5))
plt.rc('xtick', labelsize=15)
plt.xticks(rotation=90)
plt.bar(max_country,max_number1000_sorted)
plt.show()
```
